```python
import os
from ultralytics import YOLO

weights_path = "/content/drive/MyDrive/finetuned_yolo_doclaynet.pt"
kfold_root = "/content/drive/MyDrive/kfold"
```

```python
epochs = 60
batch = 16
project = "kfold_training"
results = {}
```

```python
fold_dirs = sorted([
    os.path.join(kfold_root, d)
    for d in os.listdir(kfold_root)
    if d.startswith("fold")
])
fold_dirs
```

```python
for fold_idx, fold_path in enumerate(fold_dirs):
    data_yaml = os.path.join(fold_path, "data.yaml")

    print(f"\n=== Training Fold {fold_idx} ===")
    print("Using:", data_yaml)

    model = YOLO(weights_path, task="detect")

    results[fold_idx] = model.train(
        data=data_yaml,
        epochs=epochs,
        batch=batch,
        project=project,
        imgsz=640,
        name=f"fold_{fold_idx}"
    )
```
