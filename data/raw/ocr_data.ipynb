```python
import os
from PIL import Image
import pytesseract
from tqdm import tqdm

# kalau pakai virtual env, uncomment ini, path utk tesseractnya sesuaikan dgn lokasi path tesseractmu
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
start = 508
n = 661
scan = [f'source_file/{x}.png' for x in range(start, n)]
print(len(scan), scan[0], scan[-1])
```

```python
for s in tqdm(scan, desc='Scanning images'):
    txt = pytesseract.image_to_string(Image.open(s))
    # print(txt)
    filename = f"ocr_result/ocr_{s.split('/')[-1].split('.')[0]}.txt"

    with open(filename, 'w') as file:
        file.write(txt)           
```

```python
for angka in range(start, n):
	filename = f"ground_truth/gt_{angka}.txt"

	with open(filename, 'w') as file:
		file.write("")
```
