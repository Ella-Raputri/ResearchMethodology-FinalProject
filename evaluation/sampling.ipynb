```python
import random
import os, glob, shutil
random.seed(42)
```

```python
IMG_DATASET_DIR = "../yolo/preprocess_and_training/split10/test/images"    
HEADER_DATASET_DIR =  "../data/annotated_pictures/part 5/images"
pages = []
```

```python
image_files = os.listdir(IMG_DATASET_DIR)
for filename in image_files:
    pages.append(filename.split('-')[1])

header_files = os.listdir(HEADER_DATASET_DIR)
header_files_final = []
for filename in header_files:
    header_files_final.append(filename.split('-')[1])

header_files_final.remove('378.png')
len(header_files_final)
```

```python
sampled_files = random.sample(header_files_final, 50)
pages.extend(sampled_files)
```

```python
ori_files = set(os.listdir('../data/raw/source_file'))
puretext_files = list(ori_files.difference(pages))
len(puretext_files)
```

```python
sampled_files = random.sample(puretext_files, 20)
pages.extend(sampled_files)
len(pages)
```

```python
with open('eval_list.txt','w') as file:
    file.write('\n'.join(pages))
```
